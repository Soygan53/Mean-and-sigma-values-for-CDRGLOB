```python
"""
Created on Sunday Oct 25 18:33:11 2022

@author: Ahmet Sina Soykan
"""

import pandas as pd                                           #importing libraries
import matplotlib.pyplot as plt
import numpy as np

data_all = pd.read_csv("MRI_and_CDRinfo_Values_X_train.csv")  #read this file and synchronize it "data_all"
data_labels = pd.read_csv("CDR_Values_y_train.csv")           #read this file and synchronize it "data_labels"
data_labels["HIP_VOL"] = data_all["HIPPOVOL"]                 #take "HIPPOVOL" label from data_all and add it to data_labels
A = (data_labels[data_labels["CDRGLOB"]==0].iloc[:,2])        #CDRGLOB=0 values
B = (data_labels[data_labels["CDRGLOB"]==1].iloc[:,2])        #CDRGLOB=1 values
C = (data_labels[data_labels["CDRGLOB"]==2].iloc[:,2])        #CDRGLOB=2 values
D = (data_labels[data_labels["CDRGLOB"]==3].iloc[:,2])        #CDRGLOB=3 values
A_std=A.std()                                                 #CDRGLOB=0 standart deviasion
B_std=B.std()                                                 #CDRGLOB=1 standart deviasion
C_std=C.std()                                                 #CDRGLOB=2 standart deviasion
D_std=D.std()                                                 #CDRGLOB=3 standart deviasion
A_mean=A.mean()                                               #CDRGLOB=0 mean
B_mean=B.mean()                                               #CDRGLOB=1 mean
C_mean=C.mean()                                               #CDRGLOB=2 mean
D_mean=D.mean()                                               #CDRGLOB=3 mean

bins = np.linspace(3.5, 10, 100)                        #plotting CDRGLOB=0 values, mean and sigma= -3,-2,-1,1,2,3

pyplot.hist(A, bins, alpha=0.5, label='CDRGLOB=0')
for v in [-3, -2, -1, 1, 2, 3]:
  plt.axvline(A_mean+(v*A_std), color='black')
plt.axvline(A_mean, color='red')
pyplot.legend(loc='upper right')
pyplot.show()

bins = np.linspace(2.5, 10, 100)                        #plotting CDRGLOB=1 values, mean and sigma= -3,-2,-1,1,2,3

pyplot.hist(B, bins, alpha=0.5, label='CDRGLOB=1')
for v in [-3, -2, -1, 1, 2, 3]:
  plt.axvline(B_mean+(v*B_std), color='black')
plt.axvline(B_mean, color='red')
pyplot.legend(loc='upper right')
pyplot.show()

bins = np.linspace(2.5, 8.5, 100)                     #plotting CDRGLOB=2 values, mean and sigma= -3,-2,-1,1,2,3

pyplot.hist(C, bins, alpha=0.5, label='CDRGLOB=2')        
for v in [-3, -2, -1, 1, 2, 3]:
  plt.axvline(C_mean+(v*C_std), color='black')
plt.axvline(C_mean, color='red')
pyplot.legend(loc='upper right')
pyplot.show()

bins = np.linspace(2.5, 8.5, 100)                         #plotting CDRGLOB=3 values, mean and sigma= -3,-2,-1,1,2,3

pyplot.hist(D, bins, alpha=0.5, label='CDRGLOB=3')
for v in [-3, -2, -1, 1, 2, 3]:
  plt.axvline(D_mean+(v*D_std), color='black')
plt.axvline(D_mean, color='red')
pyplot.legend(loc='upper right')
pyplot.show()
```
